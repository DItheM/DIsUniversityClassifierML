# Get Accuracy of the model

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# read data from csv files
data = pd.read_csv('train_uni_data.csv')

#prepare x and y data
x_data = data.drop(columns=['institution'])
y_data = data['institution']

#prepare and train data test data 
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
```

```python
plt.scatter(data['country'], data['institution'], label='country vs. institution')
plt.xlabel('country')
plt.ylabel('institution')
plt.legend()
plt.title('Scatter Plot')
plt.show()
```

```python
plt.scatter(data['quality'], data['institution'], label='quality vs. institution')
plt.xlabel('quality')
plt.ylabel('institution')
plt.legend()
plt.title('Scatter Plot')
plt.show()
```

```python
plt.scatter(data['budget'], data['institution'], label='budget vs. institution')
plt.xlabel('budget')
plt.ylabel('institution')
plt.legend()
plt.title('Scatter Plot')
plt.show()
```

```python
#using decision tree classifier as the model algorythmn
model = DecisionTreeClassifier()
model.fit(x_train, y_train)

#calculate accuracy
predictions = model.predict(x_test)
score = accuracy_score(y_test, predictions)
score
```

# Dump model

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib

# read data from csv files
data = pd.read_csv('train_uni_data.csv')

#prepare x and y data
x_data = data.drop(columns=['institution'])
y_data = data['institution']

#using decision tree classifier as the model algorythmn
model = DecisionTreeClassifier()
model.fit(x_data, y_data)

joblib.dump(model, 'university-classifier.pkl')

```

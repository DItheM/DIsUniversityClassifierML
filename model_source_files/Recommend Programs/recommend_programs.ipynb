# Get Accuracy of the model

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# read data from csv files
data = pd.read_csv('program_ratings.csv')

#prepare x and y data
x_data = data.drop(columns=['name'])
y_data = data['name']

#prepare and train data test data 
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
```

```python
plt.scatter(data['institution'], data['name'], label='institution vs. name')
plt.xlabel('institution')
plt.ylabel('course name')
plt.legend()
plt.title('Scatter Plot')
plt.show()
```

```python
plt.scatter(data['course_category'], data['name'], label='course_category vs. name')
plt.xlabel('rate')
plt.ylabel('course name')
plt.legend()
plt.title('Scatter Plot')
plt.show()
```

```python
plt.scatter(data['rate'], data['name'], label='rate vs. name')
plt.xlabel('rate')
plt.ylabel('course name')
plt.legend()
plt.title('Scatter Plot')
plt.show()
```

```python
#using decision tree classifier as the model algorythmn
model = DecisionTreeClassifier()
model.fit(x_train, y_train)

#calculate accuracy
predictions = model.predict(x_test)
score = accuracy_score(y_test, predictions)
score
```

# Dump model

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib

# read data from csv files
data = pd.read_csv('program_ratings.csv')

#prepare x and y data
x_data = data.drop(columns=['name'])
y_data = data['name']

#using decision tree classifier as the model algorythmn
model = DecisionTreeClassifier()
model.fit(x_data, y_data)

joblib.dump(model, 'program-recommender.pkl')

```

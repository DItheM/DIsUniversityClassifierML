```python
import pandas as pd
from sklearn.model_selection import train_test_split
from simpletransformers.seq2seq import Seq2SeqModel,Seq2SeqArgs
```

```python
model_args = Seq2SeqArgs()
model_args.num_train_epochs = 10
model_args.no_save = True
model_args.evaluate_generated_text = True
model_args.evaluate_during_training = True
model_args.evaluate_during_training_verbose = True

# Initialize model
model = Seq2SeqModel(
    encoder_decoder_type="bart",
    encoder_decoder_name="facebook/bart-large",
    args=model_args,
    use_cuda=False,
)
```

```python
# read data from csv files
data = pd.read_csv('summarize_data.csv')

#prepare x and y data
x_data = data.drop(columns=['target_text'])
y_data = data['target_text']

#prepare and train data test data 
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)

# Create DataFrames for train and test data
train_df = pd.concat([x_train, y_train], axis=1)
eval_df = pd.concat([x_test, y_test], axis=1)
```

```python
def count_matches(labels, preds):
    print(labels)
    print(preds)
    return sum(
        [
            1 if label == pred else 0
            for label, pred in zip(labels, preds)
        ]
    )
```

```python
# Train the model
model.train_model(train_df, eval_data=eval_df)
```

```python
# # Evaluate the model
results = model.eval_model(eval_df)
```

```python
import joblib
import json

model = joblib.load('feedback-summarizer.pkl')

# Open and read the JSON file
with open('feedbacks.json', 'r') as file:
    json_data = json.load(file)
    file.close
```

```python
def limitLines(summary):
    sentences = summary.split(".")
    return sentences[0]
    

def getSummary(uni_id):
    i = 0
    bucket_1 = []
    bucket_2 = []
    for x in json_data[uni_id]['feedbacks']:
        if i <= 24:
            bucket_1.append(x)
        if i >24:
            bucket_2.append(x)
        i = i + 1
        
    paragraph_1 = ' '.join(bucket_1)
    paragraph_2 = ' '.join(bucket_2)
    paragraph = paragraph_1 + ' ' + paragraph_2
    print(paragraph)
    sum_1 = limitLines(model.predict([paragraph_1])[0])
    sum_2 = limitLines(model.predict([paragraph_2])[0])
    print(sum_1 + '. ' + sum_2 + '.')
```
